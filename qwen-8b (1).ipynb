```python
from unsloth import FastLanguageModel
import torch


model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Qwen3-8B-unsloth-bnb-4bit",
    max_seq_length = 12048, 
    load_in_4bit = True
 
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 32,           
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 32,
    lora_dropout = 0, 
    bias = "none",    
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,

)
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer
import numpy as np

# Load dataset
dataset = load_dataset("billsum", split='train')  # you can also do 'test' or 'ca_test'


# Function to count tokens
def count_tokens(example):
    example['text_tokens'] = len(tokenizer.encode(example['text'], truncation=False))
    example['summary_tokens'] = len(tokenizer.encode(example['summary'], truncation=False))
    return example

# Apply function to the dataset
tokenized_dataset = dataset.map(count_tokens)

# Basic statistics
text_lengths = tokenized_dataset['text_tokens']
summary_lengths = tokenized_dataset['summary_tokens']

print("Text token length stats:")
print(f"Min: {min(text_lengths)}, Max: {max(text_lengths)}, Avg: {np.mean(text_lengths):.2f}")

print("Summary token length stats:")
print(f"Min: {min(summary_lengths)}, Max: {max(summary_lengths)}, Avg: {np.mean(summary_lengths):.2f}")
```

```python
def preprocess_text(example):
    example['text'] = example['text'].replace('\n', ' ').strip()
    example['summary'] = example['summary'].replace('\n', ' ').strip()
    return example
```

```python
legal_summarization_prompt = """Below is a legal document that needs to be summarized. Create a concise summary that captures the key points and implications of the document.

### Legal Document:
{}

### Summary:
{}"""
```

```python
EOS_TOKEN = tokenizer.eos_token

def formatting_prompts_func(examples):
    legal_documents = examples["text"]
    summaries = examples["summary"]
    
    texts = []
    for document, summary in zip(legal_documents, summaries):
        text = legal_summarization_prompt.format(document, summary) + EOS_TOKEN
        texts.append(text)
    
    return {"text": texts}
```

```python
dataset = dataset.map(preprocess_text)

dataset = dataset.map(formatting_prompts_func, batched=True)

print(dataset[0]['text'])
```

```python
from trl import SFTTrainer, SFTConfig
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset =subset ,
    args = SFTConfig(
        dataset_text_field = "text",
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4, 
        warmup_steps = 5,
        max_steps = 50,
        learning_rate = 2e-4, 
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        report_to = "wandb",
    ),
)
```

```python
trainer_stats = trainer.train()
```
